### Gridspec
and finally, but what I use the most

```python
import matplotlib.gridspec as gridspec
fig=plt.figure(figsize=[18,14])
gs = gridspec.GridSpec(2, 2,width_ratios =(0.8,0.2))


ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[0,1])
```

```python
import matplotlib.gridspec as gridspec
fig=plt.figure(figsize=[18,14])
gs = gridspec.GridSpec(2, 3)


ax1 = fig.add_subplot(gs[0,:2])
ax2 = fig.add_subplot(gs[0,2])
```

```python
import matplotlib.gridspec as gridspec
fig=plt.figure(figsize=[18,14])
gs = gridspec.GridSpec(3, 3,hspace=0.25,wspace=0.25,width_ratios =(0.6,0.2,0.2))

# hspace and wspace change the padding between plots
# width ratios changes the sizes, can also call height_ratios

ax1 = fig.add_subplot(gs[0,0])
ax2 = fig.add_subplot(gs[1,0])
ax3 = fig.add_subplot(gs[2,0])

ax4 = fig.add_subplot(gs[0,1])
ax5 = fig.add_subplot(gs[1,1])
ax6 = fig.add_subplot(gs[2,1])
```

```python
fig=plt.figure(figsize=[18,14])
gs = gridspec.GridSpec(3, 6,hspace=0.25,wspace=0.25)

# hspace and wspace change the padding between plots
# could also toggle but changing the number of components within a grid

ax1 = fig.add_subplot(gs[0,0:4])
ax2 = fig.add_subplot(gs[1,0:4])
ax3 = fig.add_subplot(gs[2,0:4])

ax4 = fig.add_subplot(gs[0,4:])
ax5 = fig.add_subplot(gs[1,4:])
ax6 = fig.add_subplot(gs[2,4:])
```

### Drawing into Axes
All plots are drawn into axes. It is easiest to understand how matplotlib works if you use the object-oriented style.

```python
# create some data to plot
import numpy as np
x = np.linspace(-np.pi, np.pi, 100)
y = np.cos(x)
z = np.sin(6*x)
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)
```

This does the same thing as

```python
plt.plot(x, y)
```

This starts to matter when we have multiple axes to worry about.

```python
fig, axes = plt.subplots(figsize=(8, 4), ncols=2)
ax0, ax1 = axes
ax0.plot(x, y)
ax1.plot(x, z)
```

#### Labeling Plots

```python
fig=plt.figure()

plt.plot(x, y)
plt.xlabel('x')
```

```python
fig, axes = plt.subplots(figsize=(8, 4), ncols=2)
ax0, ax1 = axes

ax0.plot(x, y)
ax0.set_xlabel('x')
ax0.set_ylabel('y')
ax0.set_title('x vs. y')

ax1.plot(x, z)
ax1.set_xlabel('x')
ax1.set_ylabel('z')
ax1.set_title('x vs. z')

# squeeze everything in
plt.tight_layout()
```

#### Customizing Line Plots

```python
fig, ax = plt.subplots()
ax.plot(x, y, x, z)
```

It’s simple to switch axes

```python
fig, ax = plt.subplots()
ax.plot(y, x, z, x)
```

A “parametric” graph:

```python
fig, ax = plt.subplots()
ax.plot(y, z)
```

#### Line Styles

```python
fig, axes = plt.subplots(figsize=(16, 5), ncols=3)
axes[0].plot(x, y, linestyle='dashed') ls
axes[0].plot(x, z, linestyle='--')

axes[1].plot(x, y, linestyle='dotted')
axes[1].plot(x, z, linestyle=':')

axes[2].plot(x, y, linestyle='dashdot', linewidth=5) lw
axes[2].plot(x, z, linestyle='-.', linewidth=0.5)
```

#### Colors

As described in the [colors](https://matplotlib.org/stable/api/colors_api.html) documentation, there are some special codes for commonly used colors:

- b: blue

- g: green

- r: red

- c: cyan

- m: magenta

- y: yellow

- k: black

- w: white


```python
fig, ax = plt.subplots()
ax.plot(x, y, color='k')
ax.plot(x, z, color='r')
```

Other ways to specify colors:

```python
fig, axes = plt.subplots(figsize=(16, 5), ncols=3)

# grayscale
axes[0].plot(x, y, color='0.8')
axes[0].plot(x, z, color='0.2')

# RGB tuple
axes[1].plot(x, y, color=(1, 0, 0.7))
axes[1].plot(x, z, color=(0, 0.4, 0.3))

# HTML hex code
axes[2].plot(x, y, color='#00dcba')
axes[2].plot(x, z, color='#b029ee')
```

There is a default color cycle built into matplotlib.

```python
plt.rcParams['axes.prop_cycle']
```

```python
fig, ax = plt.subplots(figsize=(12, 10))
for factor in np.linspace(0.2, 1, 11):
    ax.plot(x, factor*y)
```

#### Markers

There are lots of [different markers](https://matplotlib.org/stable/api/markers_api.html) availabile in matplotlib!

```python
fig, axes = plt.subplots(figsize=(12, 5), ncols=2)

axes[0].plot(x[:20], y[:20], marker='.')
axes[0].plot(x[:20], z[:20], marker='o')

axes[1].plot(x[:20], z[:20], marker='^',
             markersize=10, markerfacecolor='r',
             markeredgecolor='k')
```

#### Label, Ticks, and Gridlines

```python
fig, ax = plt.subplots(figsize=(12, 7))
ax.plot(x, y)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title(r'A complicated function: $f(x) = \cos(x)$')

ax.set_xticks(np.pi * np.array([-1, 0, 1]))
ax.set_xticklabels([r'$-\pi$', '0', r'$\pi$'])
ax.set_yticks([-1, 0, 1])

ax.set_yticks(np.arange(-1, 1.1, 0.2), minor=True)
#ax.set_xticks(np.arange(-3, 3.1, 0.2), minor=True)

ax.grid(which='minor', linestyle='--')
ax.grid(which='major', linewidth=2)
```

#### Axis Limits

### 2D Plotting Methods

#### imshow

```python
x1d = np.linspace(-2*np.pi, 2*np.pi, 100)
y1d = np.linspace(-np.pi, np.pi, 50)
xx, yy = np.meshgrid(x1d, y1d)
f = np.cos(xx) * np.sin(yy)
print(f.shape)
```

```python
fig, ax = plt.subplots(figsize=(12,4), ncols=2)
ax[0].imshow(f)
ax[1].imshow(f, origin='lower')
```

#### pcolormesh

```python
fig, ax = plt.subplots(ncols=2, figsize=(12, 5))
pc0 = ax[0].pcolormesh(x1d, y1d, f,shading='auto')
pc1 = ax[1].pcolormesh(xx, yy, f,shading='auto')
fig.colorbar(pc0, ax=ax[0])
fig.colorbar(pc1, ax=ax[1])
```

```python
x_sm, y_sm, f_sm = xx[:10, :10], yy[:10, :10], f[:10, :10]

fig, ax = plt.subplots(figsize=(12,5), ncols=2)

# last row and column ignored!
ax[0].pcolormesh(x_sm, y_sm, f_sm, edgecolors='k',shading='auto')

# same!
ax[1].pcolormesh(x_sm, y_sm, f_sm[:-1, :-1], edgecolors='k')
```

```python
y_distorted = y_sm*(1 + 0.1*np.cos(6*x_sm))

plt.figure(figsize=(12,6))
plt.pcolormesh(x_sm, y_distorted, f_sm[:-1, :-1], edgecolors='w',shading='flat')
plt.scatter(x_sm, y_distorted, c='k')
```

#### contour/contourf

```python
fig, ax = plt.subplots(figsize=(12, 5), ncols=2)

# same thing!
ax[0].contour(x1d, y1d, f)
ax[1].contour(xx, yy, f)
```

```python
fig, ax = plt.subplots(figsize=(12, 5), ncols=2)

c0 = ax[0].contour(xx, yy, f, 5)
c1 = ax[1].contour(xx, yy, f, 20)

plt.clabel(c0, fmt='%2.1f')  # defines decimal points on label
plt.colorbar(c1, ax=ax[1])
```

```python
fig, ax = plt.subplots(figsize=(12, 5), ncols=2)

clevels = np.arange(-1, 1, 0.2) + 0.1

cf0 = ax[0].contourf(xx, yy, f, clevels, cmap='RdBu_r', extend='both')
cf1 = ax[1].contourf(xx, yy, f, clevels, cmap='inferno', extend='both')

fig.colorbar(cf0, ax=ax[0])
fig.colorbar(cf1, ax=ax[1])
```

#### quiver

```python
u = -np.cos(xx) * np.cos(yy)
v = -np.sin(xx) * np.sin(yy)

fig, ax = plt.subplots(figsize=(12, 7))
ax.contour(xx, yy, f, clevels, cmap='RdBu_r', extend='both', zorder=0)
ax.quiver(xx[::4, ::4], yy[::4, ::4],
           u[::4, ::4], v[::4, ::4], zorder=1)
```

#### streamplot

```python
fig, ax = plt.subplots(figsize=(12, 7))
ax.streamplot(xx, yy, u, v, density=2, color=(u**2 + v**2))
```

## Handling dates on the x-axis
https://matplotlib.org/stable/api/dates_api.html

```python
# dateformatter - how you want to write the dates
# _Locator - 
```

```python
# create array of dates
import pandas as pd
xdata = pd.date_range('2018-01-01','2019-01-12',freq='W')

ydata = np.random.rand(len(xdata))
```

```python
import matplotlib.dates as mdates
myFmt1 = mdates.DateFormatter("%Y-%m-%d")    
myFmt2 = mdates.DateFormatter("%b")   
myFmt3 = mdates.DateFormatter("%Y")   


# loc = mdates.WeekdayLocator(interval=7)
# loc = mdates.YearLocator(1)
```

```python
fig= plt.figure(figsize=[10,10])


ax1 = fig.add_subplot(311)
ax1.plot(xdata,ydata)

ax2 = fig.add_subplot(312)
ax2.plot(xdata,ydata)
ax2.xaxis.set_major_formatter(myFmt2)

ax3 = fig.add_subplot(313)
ax3.plot(xdata,ydata)

# Major ticks every 6 months.
fmt_half_year = mdates.MonthLocator(interval=6)
ax3.xaxis.set_major_locator(fmt_half_year)

# Minor ticks every month.
fmt_month = mdates.MonthLocator()
ax3.xaxis.set_minor_locator(fmt_month)

#padding 



## Rotate date labels automatically
# fig.autofmt_xdate()
```
